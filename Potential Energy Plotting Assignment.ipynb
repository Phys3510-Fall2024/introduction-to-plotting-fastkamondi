### Contour Plot with Potential Energy Function
$$ U(Q_1, Q_2) = \frac{1}{2}K_1 Q_1^2 + \frac{1}{2}K_2 Q_2^2 + \frac{1}{2}C Q_1 Q_2 $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define potential energy function
def potential_energy(q1, q2 = 0, k1 = 1.0, k2 = 1.0, c = 1.0):
    return 0.5 * (k1 * q1**2 + k2 * q2**2 + c * q1 * q2)

# test function
print(potential_energy(2,2))
```

```python
# define mesh for plotting
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

X,Y = np.meshgrid(x,y)

Z = potential_energy(X,0)
```

```python
# construct contour plot
plt.figure(figsize = (3,3))
plt.contourf(X,Y,Z)

plt.title("Potential Energy of a Coupled Oscillator")

plt.xlabel("Q1 (units)")
plt.ylabel("Q2 (units)")

plt.colorbar(label = "Potential Energy (units)")

plt.show()
```
