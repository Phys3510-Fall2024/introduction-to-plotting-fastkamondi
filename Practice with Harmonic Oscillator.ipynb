```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# potential energy
def potential_energy(x, k = 1.0):
    return 0.5 * k * x**2

# test potential energy function
potential_energy(2)
```

```python
# define mesh for plotting
x = np.linspace(-5, 5, 1000) # np.linspace(x_min, x_max, npts)
y = potential_energy(x)

plt.figure(figsize = (4,3))
plt.plot(x, y, color = 'red')

# add axes labels
plt.xlabel("displacement (units)")
plt.ylabel("potential energy (units)")

# change limits of x and y
plt.xlim([-4,4])
plt.ylim([-1,10])

plt.show() # combine all elements of plot
```

### Contour Plots

$$ U(x, y) = \frac{1}{2}(k_x x^2 + k_y y^2) $$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# potential energy
def potential_energy_2D(x, y, kx = 1.0, ky = 1.0):
    return 0.5 * (kx * x**2 + ky * y**2) # kx and ky are the spring constants

# test potential energy function
potential_energy_2D(2, 0)
```

```python
# define mesh for plotting
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)

# combine 1D meshes into meshgrid
X, Y = np.meshgrid(x,y)

# define Z values
Z = potential_energy_2D(X,Y)

# print(X)
# print()
# print(Y)
# print()
# print(Z)
```

```python
# construct contour plot
plt.figure(figsize = (4,4))
plt.contourf(X,Y,Z)

# add title
plt.title("Potential Energy of a 2D Harmonic Oscillator")

# add color label
plt.colorbar(label = "potential energy (units)")

# add axes labels
plt.xlabel("x-displacement (units)")
plt.ylabel("y-displacement (units)")

# change limits of x and y
plt.xlim([-5,5])
plt.ylim([-5,5])

plt.show()
```
